```python
# Pushfront in single linkedlist 

class Node:
   def __init__(self,new_data):
      self.data = new_data
      self.next = None 

def pushfront(head,new_data):
   new_node = Node(new_data)

   new_node.next = head 

   return new_node 

def printlist(head):
   current = head 
   while current:
      print(f"{current.data}",end=' ')
      current = current.next 
   print( )

if __name__ == "__main__":
   
   head = Node(2)
   head.next = Node(3)
   head.next.next = Node(4)
   head.next.next.next = Node(5)

   print('before',end=' ')
   printlist(head)

   data = 1
   head = pushfront(head,1) 

   print("after",end=' ')
   printlist(head)
```

```python
# AddAfter in single linkedlist 

class Node:
   def __init__(self,new_data):
      self.data = new_data
      self.next = None 

def addafter(head,new_data,key):
   
   current = head
   while current:
      if current.data == key:
         break
      current = current.next 

   if current is None:
     return None
   
   new_node = Node(new_data)
   new_node.next = current.next 
   current.next = new_node

   return head


def printlist(head):
   current = head 
   while current:
      print(f"{current.data}",end=' ')
      current = current.next 
   print( )

if __name__ == "__main__":
   
   head = Node(20)
   head.next = Node(3)
   head.next.next = Node(50)
   head.next.next.next = Node(6)

   print('before',end=' ')
   printlist(head)

   data = 1
   head = addafter(head,4,3) 

   print("after",end=' ')
   printlist(head)
```

```python
# Popback in single linkedlist 

class Node:
   def __init__(self,new_data):
      self.data = new_data
      self.next = None 

def popback(head):
   
   if not head:
      return None
   
   if not head.next:
      return None 
   
   second_last = head
   while second_last.next.next:
      second_last = second_last.next
    
   second_last.next = None

   return head

def printlist(head):
   current = head 
   while current:
      print(f"{current.data}",end=' ')
      current = current.next 
   print( )

if __name__ == "__main__":
   
   head = Node(2)
   head.next = Node(3)
   head.next.next = Node(4)
   head.next.next.next = Node(5)

   print('before',end=' ')
   printlist(head)

   head = popback(head) 

   print("after",end=' ')
   printlist(head)
```

```python
# AddAfter in doubly linkedlist 

class Node:
   def __init__(self,new_data):
      self.data = new_data
      self.next = None
      self.prev = None 

def addafter_(head,new_data,key):
   
   current = head
   while current:
      if current.data == key:
         break
      current = current.next 

   if current is None:
     return head    
   
   new_node = Node(new_data)
   new_node.next = current.next 
   new_node.prev = current
   current.next = new_node

   if new_node.next:
      new_node.next.prev = new_node

   return head


def printlist(head):
   current = head 
   while current:
      print(f"{current.data}",end=' ')
      current = current.next 
   print( )

if __name__ == "__main__":
   
   head = Node(2)
   head.next = Node(3)
   head.next.prev = head
   head.next.next = Node(5)
   head.next.next.prev = head.next

   print('before',end=' ')
   printlist(head)

   data = 1
   head = addafter_(head,4,3) 

   print("after",end=' ')
   printlist(head)
```

```python
# Popfront in doubly linkedlist 

class Node:
   def __init__(self,new_data):
      self.data = new_data
      self.next = None 
      self.prev = None

def popfront(head):
   
   if not head:
      return None
   
   if not head.next:
      return None 
   
   head = head.next
   if head:
      head.prev = None

   return head

def printlist(head):
   current = head 
   while current:
      print(f"{current.data}",end=' ')
      current = current.next 
   print( )

if __name__ == "__main__":
   
   head = Node(2)
   head.next = Node(3)
   head.next.prev = head
   head.next.next = Node(4)
   head.next.next.prev = head.next

   print('before',end=' ')
   printlist(head)

   head = popfront(head) 

   print("after",end=' ')
   printlist(head)
```
